```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

#data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'], errors='coerce')
```

```python
months = df['job_posted_month'].unique()

dict_months = {month: df[df['job_posted_month'] == month] for month in months} 

df_q1 = pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index=True)
```

To csv and to excel files

```python
#exporting data
df_q1.to_csv('data_jobs_q1_2024.csv', index=False) # col=0 or index=false to avoid exporting the index
df_q1.to_excel('data_jobs_q1_2024.xlsx', index=False) # Need openpyxl installed
#conda install openpyxl
```

To_sql() 

```python
from sqlalalchemy import create_engine
engine = create_engine('sqlite:///data_jobs_q1_2024.db', echo=True) #echo true to see logs

df_q1.to_sql('jobs', con=engine, if_exists='replace', index=False)
```

to_parquet() y to_pickle() son archivos que se pueden leer mas facilmente que un csv si tienen gran cantidad de datos

```python
df_q1.to_parquet('data_jobs_q1_2024.parquet', index=False)
df_q1.to_pickle('data_jobs_q1_2024.pkl')
```
