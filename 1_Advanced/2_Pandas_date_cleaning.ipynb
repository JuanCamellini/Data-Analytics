```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data 
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df_original = dataset['train'].to_pandas()

```

```python
# median values
median_salary_hour = df['salary_hour_avg'].median()
```

```python
median_salary_year = df['salary_year_avg'].median()
```

```python
df_filled = df.copy()

df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_salary_year)
df_filled['salary_hour_avg'] = df_filled['salary_year_avg'].fillna(median_salary_hour)
```

```python
df_filled.loc[:10, 'salary_rate':'salary_hour_avg']
```

```python
df_unique = df_filled.drop_duplicates(subset=['job_title','company_name'])

#veamos cuantos valores se dropearon

print('Length before dropping duplicates:', len(df_filled))
print('Length after dropping duplicates:', len(df_unique))
print('Number of duplicates dropped:', len(df_filled) - len(df_unique))
```

```python
#Si queremos crear un dataframe que sea una copia de otro dataframe, debemos usar el metodo copy() porque sino vamos a estar trabajando con referencias al mismo objeto en memoria.
df_copy = df_unique.copy()

median_salary_hour = df_copy['salary_hour_avg'].median()

df_copy.fillna(median_salary_hour, inplace=True)
len(df)
df_copy.loc[:5, 'salary_rate':'salary_hour_avg']
```
