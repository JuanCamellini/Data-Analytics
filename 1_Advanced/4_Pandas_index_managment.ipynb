```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

#data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'], errors='coerce')
```

```python
#df.index indicates where starts and where ends the data, and the steps between them
#df.index.name = 'job_index'

""" THREE METHODS TO MANAGE THE INDEX """
#df.reset_index(inplace=True) #Resets the index to default integer index
#df.set_index('job_posted_date', inplace=True) #Sets a new index based on a column
#df.sort_index(inplace=True) #Sorts the dataframe based on the index

df_argentina = df[df['job_country'] == 'Argentina']
df_argentina.head()
#doing this the job index is not continuous anymore, so we can reset it
df_argentina.reset_index(drop=True, inplace=True)
```

```python
df_argentina.set_index('job_index', inplace=True)
#corriendo este codigo, el job_index vuelve a ser el original, y el anterior se pierde

#sort_index sirve para en casos donde el indice no tiene nombre asignado y la unica forma de ordenarlo manualmente es con sort_index
df_argentina.sort_index(inplace=True)
```
