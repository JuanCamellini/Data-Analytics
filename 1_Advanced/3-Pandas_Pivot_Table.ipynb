```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'], errors='coerce') 

```

```python
# Pivot Table 
df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median') #values es la columna que queremos agregar, index es la columna que queremos que vaya en las filas, columns es la columna que queremos que vaya en las columnas, aggfunc es la funcion de agregacion que queremos usar (en este caso la mediana)
```

```python
top_countries = df['job_country'].value_counts().nlargest(5).index
df_job_country_salary = df.pivot_table(
    values='salary_year_avg',
    index='job_country', 
    columns='job_title_short', 
    aggfunc='median'
).loc[top_countries] #.loc para filtrar las filas que queremos
job_titles = ['Data Scientist', 'Data Engineer', 'Data Analyst']

df_job_country_salary = df_job_country_salary[job_titles] #filtramos las columnas por estos tres job titles

df_job_country_salary.plot(kind='bar', figsize=(5,3))
plt.title('Median Salary by Job Title and Country')
plt.ylabel('Median Salary (Yearly)')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.legend(title='Job Title')
```
